```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000)
```

```python
d_r = 2.5 * t
```

```python
d_s = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='brown')
ax.set_xlim(0, 40)
ax.set_ylim(0, 100)
plt.xlabel('Time (t)')
plt.ylabel('Distance')
plt.title('Catching the robber')
plt.legend()
plt.grid(True)
plt.show()
```
